## one hot encoding with many categories in a feature

```python
df=pd.read_csv("mercedesbenz.csv",usecols=['X0','X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
for i in df.columns:
    print(i+' =',len(df[i].unique()) , 'Categories')
    #here we can see the categories of all the the columns.
```

```python
lst_10=df.X0.value_counts().sort_values(ascending=False).head(10).index
#here we just took only 10 most frequent categories that is present in the X0 columns
```

```python
lst_10=list(lst_10)
lst_10
```

```python
for categories in lst_10:
    df[categories]=np.where(df["X0"]==categories,1,0)
```

```python
df.head()
```

```python
lst_10.append("X0")
df[lst_10]
```
