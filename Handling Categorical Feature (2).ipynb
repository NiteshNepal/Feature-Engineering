## Count Or Frequency Encoding

```python

import pandas as pd
import numpy as np
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , header = None,index_col=None)
train_set.head() 
```

```python

columns=[1,3,5,6,7,8,9,13]
```

```python
train_set=train_set[columns]
```

```python
#we can change the name of the columns like this.
train_set.columns=['Employment','Degree','Status','Designation','family_job','Race','Sex','Country']
```

```python
train_set.head()
```

```python
for feature in train_set.columns[:]:
    print(feature,":",len(train_set[feature].unique()),'labels')
```

```python
country_map=train_set['Country'].value_counts().to_dict()
```

```python
country_map
```

```python
train_set['Country']=train_set['Country'].map(country_map)
train_set.head(20)
```

```python
####Advantages
Easy To Use
Not increasing feature space 
##### Disadvantages
It will provide same weight if the frequencies are same of different countries 
```

# Target guided ordinal Encoding 

1.we ll be ordering the labels according to the target.

2.Replace the label by the joint probability of being 1 or 0.

```python
import pandas as pd
df=pd.read_csv('titanic.csv', usecols=['Cabin','Survived'])
df.head(20)
```

```python
df['Cabin'].fillna('Missing',inplace=True)
df.head()
```

```python
#now we will be using the 1st letter of the cabin and  replace it with original.

df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean().plot.bar()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values()
```

```python
ordinal_labels=df.groupby(['Cabin'])['Survived'].mean().sort_values().index
ordinal_labels
```

```python
ordinal_labels2={k:i for i, k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
df['Cabin_ordinal_labels']=df['Cabin'].map(ordinal_labels2)
df.head(10)
```

# Mean Encoding

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)
df.head()
```

## Probability Ratio Encoding

```python
"""
1.probability of survived on cabin-----categorical feature
2.probability of not survived----1-pr(survived)
3.pr(survived)/pr(not survived)
4.Dictionary of map cabin with probabilty 
5.Replace with the categorial feature."""
```

```python
import pandas as pd
df=pd.read_csv("Titanic.csv",usecols=["Cabin","Survived"])
df.head()
```

```python
df["Cabin"].fillna("Missing",inplace=True)
df.head()
```

```python
df["Cabin"].unique()
```

```python
df["Cabin"]=df["Cabin"].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
prob_df=df.groupby(["Cabin"])["Survived"].mean().sort_values(ascending=False)
```

```python
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
prob_df["Died"]=1-prob_df["Survived"]
prob_df
```

```python
#to find the probability ratio encoding.
prob_df["Probability_ratio"]=prob_df["Survived"]/prob_df["Died"]
prob_df.head()
```

```python
probability_encoded=prob_df["Probability_ratio"].to_dict()
probability_encoded
```

```python
df['probability_encoded']=df["Cabin"].map(probability_encoded)
```
